```python
# grabing title of books with 2 star rating
import requests
import bs4
from IPython.display import clear_output
import time

l = []
s = iter(range(1,51))

res_1 = requests.get('https://toscrape.com/')
soup_1 = bs4.BeautifulSoup(res_1.text,'html.parser')

link_1 = soup_1.select('a')[0]['href']

res_2 = requests.get(link_1)
soup_2 = bs4.BeautifulSoup(res_2.text,'html.parser') 

example = soup_2.select('.product_pod')
base_url = 'https://books.toscrape.com/catalogue/page-{0}.html'

for page in range(1,51):
    base = base_url.format(page)
    res_3 = requests.get(base)
    soup_3 = bs4.BeautifulSoup(res_3.text,'html.parser')
    for i in soup_3.select('.product_pod'):
        if len(i.select('.star-rating.Two')) != 0:
            title = i.select('a')[1]
            l.append(title['title'])
    print('Page {0} complete ! '.format(next(s)))
    #time.sleep(2)
clear_output()

for i in range(len(l)):
    print('Book title : ' , l[i])
    

```

```python
base_url = 'https://books.toscrape.com/'
```

```python
list_link = example[0]('a')
```

```python
# Rating with side bar
for i in list_link:
    store_name = i['href'].split('/')
    if store_name[2] != 'books_1':
        hit = base_url + i['href']
        res_3 = requests.get(hit)
        soup_3 = bs4.BeautifulSoup(res_3.text,'html.parser')
        e = soup_3.select('.product_pod')
        print('Store : {0}'.format(store_name[3]))
        for tags in e:
            if len(tags.select('.star-rating.One')) != 0:
                print(' One Star Rating book : ')
                title = tags.select('a')[1]['title']
                print('\t Titles : {0}'.format(title))
                
            elif len(tags.select('.star-rating.Two')) != 0:
                     print(' Two Star Rating book : ')
                     title = tags.select('a')[1]['title']
                     print('\t Titles : {0}'.format(title))
                     
            elif len(tags.select('.star-rating.Three')) != 0:
                    print(' Three Star Rating book : ')
                    title = tags.select('a')[1]['title']
                    print('\t Titles : {0}'.format(title))
            
            elif len(tags.select('.star-rating.Four')) != 0:
                    print(' Four Star Rating book : ')
                    title = tags.select('a')[1]['title']
                    print('\t Titles : {0}'.format(title))
            else:
                    print(' Four Star Rating book : ')
                    title = tags.select('a')[1]['title']
                    print('\t Titles : {0}'.format(title))           
    else:
        pass
```
